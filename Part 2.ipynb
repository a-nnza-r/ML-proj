```python
# Open the file
with open("Data/ES/train", "r", encoding="UTF-8") as file:
    # Read the entire contents of the file
    training_contents = file.read().split("\n")
    
```

```python
# Extract all the lables in a particular order, leaving the spaces as label "Space"
labels = []
for line in training_contents:
    split_line = line.split(" ")
    if len(split_line) > 1:
        labels.append(split_line[1])
    else:
        labels.append("STOP")
        labels.append("START")
        
# Adding START to the first word and removing START from the last word and removing repeat START/STOPS
labels.insert(0, "START")
labels.pop()
labels.pop()
labels.pop()


        
    
```

```python
print(labels[-1])
```

```python
# Now to calculate the transition probabiltiies of states START, (given 7 states), STOP
transition_probabilties = {}

for i in range(len(labels)):
    if i < len(labels)-1:
        if labels[i] not in transition_probabilties:
            transition_probabilties[labels[i]] = {}
        
        # Check the next word and add it as a count
        if labels[i+1] not in transition_probabilties[labels[i]]:
            transition_probabilties[labels[i]][labels[i+1]] = 1
        else:
            transition_probabilties[labels[i]][labels[i+1]] += 1
```

```python
transition_probabilties
```

```python
# Calculate the transition probabilities from above
for outside_label in transition_probabilties:
    total_count = sum(transition_probabilties[outside_label].values())
    for inside_label in transition_probabilties[outside_label]:
        print(transition_probabilties[outside_label][inside_label], total_count)
        transition_probabilties[outside_label][inside_label] = transition_probabilties[outside_label][inside_label] / total_count
        
```
